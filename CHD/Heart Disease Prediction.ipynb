## Importing

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('framingham_heart_disease.csv')
df.head()
```

```python
print('No. of coloumns : ', len(df.columns))
print('No. of rows : ', df.shape[0])
```

```python
print('Is Na :\n', df.isna().sum(), '\n')
print('Total isNa : ', df.isna().sum().sum())
```

```python
df_clean = df.dropna(axis = 0)
df_clean.head()
```

```python
print('Is Na :\n', df_clean.isna().sum(), '\n')
print('Total isNa : ', df_clean.isna().sum().sum())
```

```python
print(df_clean.shape)
```

```python
Y = df_clean.filter(['TenYearCHD'])
Y
```

```python
X = df_clean.drop(['TenYearCHD'], axis = 1)
X = X.reset_index(drop = True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, shuffle = False)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
reg = lr.fit(X_train, Y_train)
Y_pred = reg.predict(X_test)
```

```python
from sklearn import metrics
print('Mean Absolute Error : ', metrics.mean_absolute_error(Y_test, Y_pred)*100, '%')
print('Mean Squared Error : ', metrics.mean_squared_error(Y_test, Y_pred)*100, '%')
print('Root Mean Squared Error : ', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred))*100, '%')
```
